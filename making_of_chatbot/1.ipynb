````python

%%writefile app.py
import streamlit as st

with st.chat_message('user'): #this will tell who is it user or ai
  st.text('Hi') # this will create the user box
with st.chat_message('ai'): #this will tell who is it user or ai
  st.text('How can i help u')
# we will create the input variable where user will type his or her messgae
user_input = st.chat_input('Enter your message')
# if u only do this u will press enter and it will not be shown in text messgae
# it is just we because we have not define action
# ----- action if user type sometime ----
if user_input:
  with st.chat_message('user'):
    st.text(user_input) #text box mein ky display honga jo bhi user_input hai
  with st.chat_message('ai'):
    st.text('How can i help u')
````
